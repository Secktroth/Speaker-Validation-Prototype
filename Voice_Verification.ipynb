```python
# --- Install SpeechBrain (run once per fresh runtime) ---
!pip -q install --upgrade pip
!pip -q install speechbrain

# --- Patch torchaudio BEFORE importing SpeechBrain --- -> doing this because of weird errors arising earlier
import torchaudio
if not hasattr(torchaudio, "list_audio_backends"):
    def list_audio_backends():
        return ["torchcodec"]
    torchaudio.list_audio_backends = list_audio_backends

# --- Now import SpeechBrain ---
from speechbrain.pretrained import SpeakerRecognition

# --- Mount Drive ---
from google.colab import drive
drive.mount("/content/drive")

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# --- Paths for Case --- -> enter google drive file paths here
ref_path = Path("")
testing_path = Path("")

# Allowing different formats
audio_exts = {".wav", ".mp3", ".flac", ".m4a", ".ogg", ".aac", ".wma"}

# --- Pick the single reference file ---
ref_files = [p for p in ref_path.iterdir() if p.is_file() and p.suffix.lower() in audio_exts]
if len(ref_files) == 0:
    raise FileNotFoundError(f"No audio file found in reference folder: {ref_path}")
if len(ref_files) > 1:
    print("Warning: more than one reference audio file found. Using the first:")
    for p in ref_files:
        print(" -", p.name)
reference_file = ref_files[0] # using the first and only ref file (this 0 is referring to the position in the array)

# --- Collect test files ---
test_files = sorted([p for p in testing_path.iterdir() if p.is_file() and p.suffix.lower() in audio_exts])
if len(test_files) == 0:
    raise FileNotFoundError(f"No audio files found in testing folder: {testing_path}")

print("Reference:", reference_file.name)
print("Tests:", [p.name for p in test_files])

# --- Load verifier ---
verifier = SpeakerRecognition.from_hparams(
    source="speechbrain/spkrec-ecapa-voxceleb",
    savedir="/content/pretrained_models/spkrec-ecapa-voxceleb",
)

# --- Verify ---
threshold = 0.30
rows = []

for tf in test_files:
    score, _ = verifier.verify_files(str(reference_file), str(tf))
    score = float(score.item() if hasattr(score, "item") else score)
    rows.append({
        "test_file": tf.name,
        "score": score,
        "prediction": "Same Speaker" if score >= threshold else "Different Speakers",
    })

df = pd.DataFrame(rows).sort_values("score", ascending=False).reset_index(drop=True)
display(df)

# --- Plot ---
plt.figure(figsize=(10, 4))
plt.bar(df["test_file"], df["score"])
plt.axhline(
    y=threshold,
    linestyle="--",
    color="orange",
    label=f"Threshold ({threshold})"
)
plt.xticks(rotation=25, ha="right")
plt.ylabel("Score")
plt.title("SpeechBrain ECAPA: Reference vs Files to Test")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# --- Install dependencies ---
!pip -q install resemblyzer scikit-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from google.colab import drive

from resemblyzer import VoiceEncoder, preprocess_wav
from sklearn.metrics.pairwise import cosine_similarity

# --- Mount Google Drive ---
drive.mount("/content/drive")

# --- Paths for Case --- -> enter google drive file paths here
ref_path = Path("")
testing_path = Path("")

print("Reference folder exists:", ref_path.exists())
print("Testing folder exists:", testing_path.exists())

audio_exts = {".wav", ".mp3", ".flac", ".m4a", ".ogg", ".aac", ".wma"}

# --- Get the single reference file ---
ref_files = [p for p in ref_path.iterdir() if p.is_file() and p.suffix.lower() in audio_exts]
if len(ref_files) == 0:
    raise FileNotFoundError(f"No audio file found in reference folder: {ref_path}")
if len(ref_files) > 1:
    print("Warning: more than one reference audio file found. Using the first:")
    for p in ref_files:
        print(" -", p.name)

reference_file = ref_files[0]
print("\nUsing reference file:", reference_file.name)

# --- Get test files ---
test_files = sorted([p for p in testing_path.iterdir() if p.is_file() and p.suffix.lower() in audio_exts])
if len(test_files) == 0:
    raise FileNotFoundError(f"No audio files found in testing folder: {testing_path}")

print("\nTesting files:")
for p in test_files:
    print(" -", p.name)

# --- Initialise encoder once ---
encoder = VoiceEncoder()

def get_voice_embedding(audio_file: Path) -> np.ndarray:
    """Load and compute the voice embedding for a given audio file."""
    wav = preprocess_wav(str(audio_file))
    return encoder.embed_utterance(wav)

# --- Compute reference embedding once ---
embedding_ref = get_voice_embedding(reference_file)

# --- Similarity threshold ---
# Note: Resemblyzer similarity is typically high for same-speaker, but exact values vary by audio conditions.
threshold = 0.75

rows = []
for tf in test_files:
    embedding_test = get_voice_embedding(tf)
    similarity = cosine_similarity([embedding_ref], [embedding_test])[0][0]
    rows.append({
        "test_file": tf.name,
        "cosine_similarity": float(similarity),
        "similarity_%": float(similarity * 100),
        "prediction": "Same Speaker" if similarity >= threshold else "Different Speakers"
    })

df = pd.DataFrame(rows).sort_values("cosine_similarity", ascending=False).reset_index(drop=True)

print("\n--- Results ---")
print("Reference:", reference_file.name)
print("Threshold:", threshold)
display(df)

# --- Plot (percent) ---
plt.figure(figsize=(10, 4))
plt.bar(df["test_file"], df["similarity_%"])
plt.axhline(
    y=threshold * 100,
    linestyle="--",
    color="orange",
    label=f"Threshold ({threshold*100:.0f}%)"
)
plt.xticks(rotation=25, ha="right")
plt.ylabel("Similarity (%)")
plt.ylim(0, 100)
plt.title("Resemblyzer Voice Similarity: Reference vs Test Files (Cosine)")
plt.legend()
plt.tight_layout()
plt.show()

```
